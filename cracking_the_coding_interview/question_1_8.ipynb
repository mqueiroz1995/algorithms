# Question 1.8

**Zero Matrix:** Write an algorithm such that if an element in an MxN matrix is 0, its entire row and column are set to 0.

## Examples

```
A:
  [1 2 3]    [1 0 3]
  [1 0 3] -> [0 0 0]
  [1 2 3]    [1 0 3]

B:
  [0 2 3]    [0 0 0]
  [1 2 3] -> [0 2 3]
  [1 2 3]    [0 2 3]
```

## Constraints

* What is the expected output for an empty matrix?

## Ideas

### Idea 1

Iterate throw whole matrix and store rows and columns that should be zeroed in memory.

Iterate again throw whole matrix transforming positions that should be zeroed into zero.

**Time**: O(n*m) - memory is O(n + m)
**Space**: O(n*m)

````python
# ALGORITHM
def process_array(a):
    z_colums = set()
    z_rows = set()

    for i, _ in enumerate(a):
        for j, _ in enumerate(a[i]):
            if a[i][j] == 0:
                z_colums.add(i)
                z_rows.add(j)

    for i, _ in enumerate(a):
        for j, _ in enumerate(a[i]):
            if i in z_colums or j in z_rows:
                a[i][j] = 0

    return a

# TESTS
def test(input_a, expected_response):
    response = process_array(input_a)
    if not array_equal(response, expected_response):
        print("Test fails for input\n'{}'.\nExpected value:\n{}\nReturned value:\n{}\n".format(input_a, expected_response, response))
    else:
        print("Test passes for input\n'{}'.\n".format(input_a))

test([[]], [[]])


input = ([
    [1, 2, 3],
    [4, 0, 6],
    [7, 8, 9]
])

expected_response = ([
    [1, 0, 3],
    [0, 0, 0],
    [7, 0, 9]
])

test(input, expected_response)


input = ([
    [0, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
])

expected_response = ([
    [0, 0, 0],
    [0, 5, 6],
    [0, 8, 9]
])

test(input, expected_response)
````
